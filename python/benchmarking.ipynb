## PyRoaring vs Splinters

This notebook analyzes the output of automated memory-footprint benchmarks comparing the PyRoaring python package (which wraps the CRoaring implementation of Roaring Bitmaps) with the splynters python package, which wraps the splinter-rs bitmap compression algorithm, created by Carl Sverre.

To run this benchmark locally, ensure that you have all relevant datasets in the real_roaring_data directory, then from the root directory of this repository run `python3 python/run_all_benchmarks.py`. The results of the benchmarks will be written to dataframes in the benchmark_results directory, which this notebook will visualize.

Each set of files are analyzed in terms of the overall compressed bitmap size, compression ratio, and compression advantage of splinter/roaring: an advantage > 1 indicates that Splinter compresses the file further, and advantage < 1 indicates that PyRoaring compresses the file further.

All datasets used in these benchmarks are gathered from the Real Roaring Datasets benchmarks directory: https://github.com/RoaringBitmap/real-roaring-datasets/tree/master


```python
df = pd.read_csv("../benchmark_results/results_dimension_003.csv")

unc = df["Uncompressed (bytes)"]
roaring = df["Roaring (bytes)"]
splinter = df["Splinter (bytes)"]
roaring_r = df["Roaring Ratio"]
splinter_r = df["Splinter Ratio"]

plt.scatter(unc, roaring, label = "Roaring")
plt.scatter(unc, splinter, label = "Splinter")
plt.xlabel("Uncompressed size (bytes)")
plt.ylabel("Compressed size (bytes)")
plt.legend()
plt.title("Roaring vs Splinter Compression Benchmarks (Dimension_003)")
plt.axline((0, 0), slope=1)
plt.show()

plt.scatter(unc, roaring_r, label = "Roaring")
plt.scatter(unc, splinter_r, label = "Splinter")
plt.xlabel("Uncompressed size (bytes)")
plt.ylabel("Compression ratio")
plt.legend()
plt.title("Roaring vs Splinter Compression Ratios (Dimension_003)")
plt.axline((0, 1), slope=0)
plt.show()

plt.scatter(unc, splinter_r/roaring_r)
plt.xlabel("Uncompressed size (bytes)")
plt.ylabel("Compression advantage (splinter/roaring")
plt.title("Roaring vs Splinter Compression Advantage (Dimension_003)")
plt.axline((0, 1), slope=0)
plt.show()
```
